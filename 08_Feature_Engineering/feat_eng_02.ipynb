```python
# Let's Check the how many Questions marks in the data

columns = ['age', 'workclass', 'fnlwgt', 'education', 'education.num',
       'marital.status', 'occupation', 'relationship', 'race', 'sex',
       'capital.gain', 'capital.loss', 'hours.per.week', 'native.country',
       'income']

count = 0

for col in columns :
    for val in df[col]:
        if (val == "?"):
            # print("hey this is here")
            count = count + 1

print(count)
```

```python
# Get The all Integer Columns 

data_col = df.select_dtypes(include="int64").columns
```

```python
# Fill the null values with Simple Imputer 

from sklearn.impute import SimpleImputer
mean_val = SimpleImputer(strategy = "mean")

for col in data_col:
    df[col] = mean_val.fit_transform(df[[col]])    
    print(col)
```
